```python
# Autoencoders
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import regularizers

# Load the DataFrame
# df = pd.read_csv("C:/Users/44743/Desktop/Data Science/IPL Data Set/DataSet/team_performance_dataset_2008to2024.csv")

# Select relevant columns
relevant_columns = ['Venue', 'Toss_Winner', 'Toss_Decision', 'First_Innings_Score', 'Powerplay_Scores']

# Filter data for RCB matches
rcb_matches = df[df['Teams'].str.contains('Royal Challengers Bangalore')]

# Encode categorical variables (e.g., one-hot encoding)
X = pd.get_dummies(rcb_matches[relevant_columns])

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, _, _ = train_test_split(X_scaled, X_scaled, test_size=0.2, random_state=42)  # Using X_scaled for both X_train and y_train

# Initialize autoencoder model
autoencoder = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(64, activation='relu'),
    Dense(128, activation='relu'),
    Dense(X_train.shape[1])
])

# Compile the autoencoder model
autoencoder.compile(optimizer='adam', loss='mse')

# Train the autoencoder model
autoencoder.fit(X_train, X_train, epochs=100, batch_size=32)

# Extract features from autoencoder's middle layer
encoder = Sequential(autoencoder.layers[:-3])

# Extracting features from autoencoder
X_train_encoded = encoder.predict(X_train)
X_test_encoded = encoder.predict(X_test)

# Initialize ANN model with reduced features
model = Sequential([
    Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01), input_shape=(X_train_encoded.shape[1],)),
    Dropout(0.5),  # Dropout rate of 0.5 (50% dropout)
    Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compile the ANN model
model.compile(optimizer='RMSprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the ANN model with early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
model.fit(X_train_encoded, y_train, epochs=400, batch_size=32, validation_data=(X_test_encoded, y_test), callbacks=[early_stopping])

# Evaluate the ANN model
accuracy = model.evaluate(X_test_encoded, y_test)[1]
print("Accuracy:", accuracy)

# Use the trained ANN model to predict future matches for RCB
# You can input new data and make predictions using model.predict(new_data)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import MiniBatchDictionaryLearning
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import regularizers

# Load the DataFrame
df = pd.read_csv("C:/Users/44743/Desktop/Data Science/IPL Data Set/DataSet/team_performance_dataset_2008to2024.csv")

# Select relevant columns
relevant_columns = ['Venue', 'Toss_Winner', 'Toss_Decision', 'First_Innings_Score', 'Powerplay_Scores']

# Filter data for RCB matches
rcb_matches = df[df['Teams'].str.contains('Royal Challengers Bangalore')]

# Encode categorical variables (e.g., one-hot encoding)
X = pd.get_dummies(rcb_matches[relevant_columns])

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)

# Initialize MiniBatchDictionaryLearning model
sparse_coder = MiniBatchDictionaryLearning(n_components=32, alpha=1, batch_size=100, n_iter=100)

# Fit sparse coder to training data
sparse_coder.fit(X_train)

# Transform training and testing data to sparse representations
X_train_sparse = sparse_coder.transform(X_train)
X_test_sparse = sparse_coder.transform(X_test)

# Build the ANN model
model = Sequential([
    Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01), input_shape=(X_train_sparse.shape[1],)),
    Dropout(0.5),  # Dropout rate of 0.5 (50% dropout)
    Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.01)),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compile the ANN model
model.compile(optimizer='RMSprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the ANN model with early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
model.fit(X_train_sparse, y_train, epochs=400, batch_size=32, validation_data=(X_test_sparse, y_test), callbacks=[early_stopping])

# Evaluate the ANN model
accuracy = model.evaluate(X_test_sparse, y_test)[1]
print("Accuracy:", accuracy)

# Save the trained model
model.save("rcb_match_prediction_model.h5")

# Load the DataFrame for new data
df_new = pd.read_csv("C:/Users/44743/Desktop/Data Science/IPL Data Set/DataSet/team_performance_dataset_RCB.csv")

# Select relevant columns for new data
relevant_columns_new = ['Venue', 'Toss_Winner', 'Toss_Decision', 'First_Innings_Score', 'Powerplay_Scores']

# Filter data for RCB matches
rcb_matches_live = df_new[df_new['Teams'].str.contains('Royal Challengers Bangalore')]

# Preprocess the new data
new_X = pd.get_dummies(rcb_matches_live[relevant_columns_new], drop_first=True)  # Encode categorical variables

# Ensure that the new data has the same columns as the training data after one-hot encoding
missing_columns = set(X.columns) - set(new_X.columns)
for column in missing_columns:
    new_X[column] = 0  # Add missing columns with default value 0

new_X = new_X[X.columns]  # Reorder columns to match the order in the training data

new_X_scaled = scaler.transform(new_X)  # Standardize features
new_X_sparse = sparse_coder.transform(new_X_scaled)  # Transform data to sparse representations

# Make predictions
predictions = model.predict(new_X_sparse)

# Convert probabilities to binary labels
binary_predictions = (predictions > 0.5).astype(int)

print(binary_predictions)
print(binary_predictions.sum())
```
